```python
import pandas as pd
from urllib.request import urlopen
from geopy.geocoders import Nominatim
import numpy as np
# import json
import json
```

```python
# store the URL in url as
# parameter for urlopen
url = "https://api.geonet.org.nz/intensity/strong/processed/2020p666015"

# store the response of URL
response = urlopen(url)
```

```python
# storing the JSON response
# from url in data

data_json = json.loads(response.read())
intensity_raw_df = pd.json_normalize(data_json["features"])
# intensity_raw_df
```

```python
## Split geometry.coordinates into longitude and latitude

intensity_raw_df = intensity_raw_df.astype({'geometry.coordinates':'string'})
intensity_raw_df

# Split the "geometry.coordinates" column into "longitude" and "latitude" columns
intensity_raw_df['longitude'] = intensity_raw_df['geometry.coordinates'].str.split(',').str[0].str.strip('[ ')
intensity_raw_df['latitude'] = intensity_raw_df['geometry.coordinates'].str.split(',').str[1].str.strip('] ')

# Convert the columns to the appropriate data type if needed
intensity_raw_df['longitude'] = intensity_raw_df['longitude'].astype(float)
intensity_raw_df['latitude'] = intensity_raw_df['latitude'].astype(float)

intensity_raw_df
```

```python
# stage to additional clean, cut , convert and rename columns
quake_intensity_df = intensity_raw_df
quake_intensity_df.head()
```

```python
## export to CSV
quake_intensity_df.to_csv('quake_intensity_data.csv', encoding='utf-8')
```
