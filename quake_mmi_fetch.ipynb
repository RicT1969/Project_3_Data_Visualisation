```python
import pandas as pd
from urllib.request import urlopen
# from geopy.geocoders import Nominatim
import numpy as np
import os
import csv
import datetime as dt


# import json
import json
# store the URL in url as 
# parameter for urlopen

## API URL  - using quakes MMI "2" - "7"
url2 = "https://api.geonet.org.nz/quake?MMI=2"
url3 = "https://api.geonet.org.nz/quake?MMI=3"
url4 = "https://api.geonet.org.nz/quake?MMI=4"
url5 = "https://api.geonet.org.nz/quake?MMI=5"
url6 = "https://api.geonet.org.nz/quake?MMI=6"
url7 = "https://api.geonet.org.nz/quake?MMI=7"
  
```

```python
## Create DF for each API request at varying levels of MMI 
# ---------------------------
# MMI = 2
# ---------------------------
# store the response of URL
response2 = urlopen(url2)
  
# storing the JSON response 
# from url in data
data_json2 = json.loads(response2.read())

MMI2_df = pd.json_normalize(data_json2["features"])
# MMI2_df.head()

# ---------------------------
# MMI = 3
# ---------------------------
# store the response of URL
response3 = urlopen(url3)
  
# storing the JSON response 
# from url in data
data_json3 = json.loads(response3.read())

MMI3_df = pd.json_normalize(data_json3["features"])
# MMI2_df.head()

# ---------------------------
# MMI = 4
# ---------------------------
# store the response of URL
response4 = urlopen(url4)
  
# storing the JSON response 
# from url in data
data_json4 = json.loads(response4.read())

MMI4_df = pd.json_normalize(data_json4["features"])
# MMI4_df.head()


# ---------------------------
# MMI = 5
# ---------------------------
# store the response of URL
response5 = urlopen(url5)
  
# storing the JSON response 
# from url in data
data_json5 = json.loads(response5.read())

MMI5_df = pd.json_normalize(data_json5["features"])
# MMI5_df.head()

# ---------------------------
# MMI = 6
# ---------------------------
# store the response of URL
response6 = urlopen(url6)
  
# storing the JSON response 
# from url in data
data_json6 = json.loads(response6.read())

MMI6_df = pd.json_normalize(data_json6["features"])
# MMI6_df.head()

# ---------------------------
# MMI = 7
# ---------------------------
# store the response of URL
response7 = urlopen(url7)
  
# storing the JSON response 
# from url in data
data_json7 = json.loads(response7.read())

MMI7_df = pd.json_normalize(data_json7["features"])
# MMI7_df.head()
```

```python
# Union raw all dataframes via concat
# Put into list then concat.

MMI_raw_df_list = [MMI2_df,MMI3_df,MMI4_df,MMI5_df,MMI6_df,MMI7_df]
MMI_union_df = pd.concat(MMI_raw_df_list)
# union_raw_df.count()

MMI_union_df.head()
```

```python
## Split geometry.coordinates into longitude and latitude

MMI_union_df = MMI_union_df.astype({'geometry.coordinates':'string'})
MMI_union_df

# Split the "geometry.coordinates" column into "longitude" and "latitude" columns
MMI_union_df['longitude'] = MMI_union_df['geometry.coordinates'].str.split(',').str[0].str.strip('[ ')
MMI_union_df['latitude'] = MMI_union_df['geometry.coordinates'].str.split(',').str[1].str.strip('] ')

# Convert the columns to the appropriate data type if needed
MMI_union_df['longitude'] = MMI_union_df['longitude'].astype(float)
MMI_union_df['latitude'] = MMI_union_df['latitude'].astype(float)

# DF below combines all API Requests however is still raw from JSON.
MMI_union_df.head()
```

```python
## Final Clean and convert
## stage to additional clean, cut , convert and rename columns
## Check Index
# MMI_union_df.columns.get_loc("properties.publicID")

# Rename Columns
quake_mapping = {MMI_union_df.columns[3]:'publicID',MMI_union_df.columns[4]:'time',MMI_union_df.columns[5]:'depth',MMI_union_df.columns[6]:'magnitude',MMI_union_df.columns[7]:'mmi',MMI_union_df.columns[8]:'locality'}
MMI_new_column_df = MMI_union_df.rename(columns =quake_mapping)

#Create date column 
MMI_new_column_df['date'] =pd.to_datetime(MMI_new_column_df['time']).dt.strftime('%Y-%m-%d')

# strip back to columns needed.
MMI_new_clean_df = MMI_new_column_df[["publicID","date","mmi","magnitude","locality","longitude","latitude","time","depth"]]


# Convert into final data frame quake_mmi_df
quake_mmi_df = pd.DataFrame(MMI_new_clean_df)
quake_mmi_df.reset_index
quake_mmi_df.head()
```

```python
## export to CSV
quake_mmi_df.to_csv('quake_mmi_data.csv', encoding='utf-8')
```
