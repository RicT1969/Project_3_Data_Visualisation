```python
import pandas as pd
from urllib.request import urlopen
from geopy.geocoders import Nominatim
import numpy as np
# import json
import json
```

```python
# store the URL in url as
# parameter for urlopen

# "volcano" url
url = "https://api.geonet.org.nz/volcano/val"

# store the response of URL
response = urlopen(url)
```

```python
# storing the JSON response
# from url in data

data_json = json.loads(response.read())
volcano_raw_df = pd.json_normalize(data_json["features"])
volcano_raw_df
```

```python
## Split geometry.coordinates into longitude and latitude

volcano_raw_df = volcano_raw_df.astype({'geometry.coordinates':'string'})
volcano_raw_df

# Split the "geometry.coordinates" column into "longitude" and "latitude" columns
volcano_raw_df['longitude'] = volcano_raw_df['geometry.coordinates'].str.split(',').str[0].str.strip('[ ')
volcano_raw_df['latitude'] = volcano_raw_df['geometry.coordinates'].str.split(',').str[1].str.strip('] ')

# Convert the columns to the appropriate data type if needed
volcano_raw_df['longitude'] = volcano_raw_df['longitude'].astype(float)
volcano_raw_df['latitude'] = volcano_raw_df['latitude'].astype(float)

volcano_raw_df
```

```python
# Final clean and relabel of columsn

# Rename Columns
volcano_mapping = {volcano_raw_df.columns[3]:'acc',volcano_raw_df.columns[4]:'activity',volcano_raw_df.columns[5]:'hazards', volcano_raw_df.columns[6]:'level',volcano_raw_df.columns[7]:'volcanoID',volcano_raw_df.columns[8]:'volcanoTitle'}
volcano_raw_df = volcano_raw_df.rename(columns =volcano_mapping)
# intensity_raw_df

# Strip columsn for final DF
volcano_data_df = pd.DataFrame(volcano_raw_df[['volcanoID','volcanoTitle','longitude','latitude','acc','activity','level','hazards']])
volcano_data_df.head()

```

```python
## export to CSV without index
volcano_data_df.to_csv('volcano_data.csv', encoding='utf-8',index=False)
```
